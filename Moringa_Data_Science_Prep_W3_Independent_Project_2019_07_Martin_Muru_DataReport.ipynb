# Data Cleaning and exploration 

Notice that some column names in the 3 dataframes have different names. Renaming column names so that they match in all the 3 dataframes before concatinatiing them

```python
telcom1.rename(columns = {'PRODUTC':'PRODUCT','DATETIME':'DATE_TIME'}, inplace = True)
```

```python
telcom2.rename(columns = {'DW_A_NUMBER':'DW_A_NUMBER_INT','DW_B_NUMBER':'DW_B_NUMBER_INT'}, inplace = True)
```

```python
telcom3.rename(columns = {'CELLID':'CELL_ID','SIET_ID':'SITE_ID'}, inplace = True)
```

```python
telcom1.head(1)
```

```python
telcom2.head(1)
```

```python
telcom3.head(1)
```

Concatinating the 3 dataframes to form one telcom_data dataframe

```python
telcom_data = pd.concat([telcom1,telcom2,telcom3])
telcom_data
```

Joining the geodata dataframe to the telecom dataframe. To do this I have to find the Column that has matching keys in the two dataframes

```python
geodata.head(5)
```

```python
telcom_data.head(5)
```

```python
#Saving Cell_id column from geodata as a List
Geocelllist = geodata['CELL_ID'].tolist()
```

```python
#Saving Cell_id column from telecom as a List

TelcomList =telcom_data['CELL_ID'].tolist()
```

```python
#Checking if they have any matching Keys
set(Geocelllist) & set(TelcomList)
```

Cellid has no matching keys. 
Checking for matching keys on SiteID

```python
# Checking for matching keys from Site Id
Geositelist = geodata['SITE_CODE'].tolist()
TelcomList =telcom_data['SITE_ID'].tolist()
```

```python
set(Geositelist) & set(TelcomList)
```

There exists matching keys on site id. We will use this to merge the geodata dataframe to the Telcom dataframe

```python
#Renaming Site code to site id
geodata.rename(columns = {'SITE_CODE':'SITE_ID'}, inplace = True)
```

```python
#Merging the two dataframes using Site ID
mtn_df= pd.merge(geodata, telcom_data, on='SITE_ID', how='inner')
```

```python
mtn_df
```

# Data Cleaning

```python
#Checking for duplicates
mtn_df[mtn_df.duplicated()]
```

```python
#Dropping the duplicates
mtn_df = mtn_df.drop_duplicates()
```

```python
#Making sure they are dropped
mtn_df[mtn_df.duplicated()]
```

```python
#checking for null Values
mtn_df[mtn_df.isnull()].any(axis=1)
```

```python
#Dropping Null Values
mtn_df = mtn_df.dropna(how='any')
```

```python
#checking that null values have been dropped
mtn_df[mtn_df.isnull()].any(axis = 1)
```

```python
#Checking for information on the data such as datatype
mtn_df.info()
```

```python
#Date time column is saved as a string. Converting it to DATETIME below:
mtn_df['DATE_TIME'] = pd.to_datetime(mtn_df['DATE_TIME'])
```

```python
mtn_df.info()
```

```python
#Dropping Country Columns since its the same country
mtn_df = mtn_df.drop(['COUNTRY_A','COUNTRY_B'], axis = 1)
```

```python
mtn_df.head(2)
```

# Data Exploration/Analysis

```python
#Getting some useful statistics about the string objects in the data
mtn_df.describe(include = 'object')
```

```python
#Creating a column Day name to find the days where MTN Infrastructure is used the most
mtn_df['Day_name'] = mtn_df['DATE_TIME'].dt.day_name()
```

```python
#Creating a column Hour to find the hours when mtn Infrastucture is used the most
mtn_df['Hour'] = mtn_df['DATE_TIME'].dt.hour
```

```python
mtn_df.head(10)
```

```python
mtn_df['Day_name'].value_counts()
```

From the data, Most users use MTN Services on Wednesday. If they are to upgrade their infrastructure, Wednesday would not be ideal since alot of people are using it. A day like Tuesday that isnt as busy would be a better day.

Which were the most used cities in the three days?

```python
mtn_df['VILLES'].value_counts()
```

YOPOUGON, Abobo and Cocody are the top 3 cities where alot of people use the MTN Services. These should be of Priority. 

```python
#Creating a column that will specify whether an MTN Service was used during home hours or Work hours. 
#Assumption : Work hours is from 8am to 5pm and home hours is the rest of the day.
mtn_df['Hour_desc'] = mtn_df['Hour'].apply(lambda x: 'Work Hours' if x >= 8 and x <=17 else 'Home Hours') 
```

```python
mtn_df['Hour_desc'].unique()
```

It seems that the dataset only has data about Home hours. We would require data for between 8am to 5pm to be able to accurately determine the times when MTN Services are mostly used. This data is missing.

```python
#Determining the service mostly used in the dataset
mtn_df['PRODUCT'].value_counts()
```

Most people were using SMS To communicate. Further analysis and data is required to know why this is the case. Are the voice services too expensive? Are they unreliable and hence urgently require infrastructure repair?Lets find out the average billing Price for each

```python
mtn_df.groupby('PRODUCT')['VALUE'].mean()
```

Voice Services are way more expensive on average than SMS Services. This somewhat explains why SMS is the more used service

Some cities bring as low as 0 returns in terms of billing. This might be areas MTN can concentrate on improving infrastructure on.
